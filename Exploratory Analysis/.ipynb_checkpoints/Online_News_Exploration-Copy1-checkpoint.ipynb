```python
#Figure out share buckets
shares_df = filtered_online_news_df['Shares_per_day']
filtered_online_news_df['Shares_per_day'].describe()
```

```python
new_filtered_online_news_df = filtered_online_news_df[filtered_online_news_df["Shares_per_day"] <6000]
```

```python
new_shares_df = new_filtered_online_news_df['Shares_per_day']
new_filtered_online_news_df['Shares_per_day'].describe()
```

```python
Share_bins = [-1, 14, 5265]
labels = ['Less Popular','Highly Popular']
new_filtered_online_news_df["Popularity"] = pd.cut(new_filtered_online_news_df["Shares_per_day"], Share_bins, labels=labels)
new_filtered_online_news_df.head(10)
```

```python
ax = new_filtered_online_news_df['Popularity'].value_counts().plot(kind='bar', figsize=(14,8), title="Article Count by Popularity")
ax.set_xlabel("Popularity Bin")
ax.set_ylabel("Article Counts")
```

```python
online_news_df_encoded = pd.get_dummies(new_filtered_online_news_df, columns=["Popularity"])
online_news_df_encoded.head()
```

```python
X = online_news_df_encoded.drop([" timedelta","Popularity_Highly Popular","Popularity_Less Popular"," shares", "Shares_per_day"], axis=1)
X.head()
```

```python
y = online_news_df_encoded["Popularity_Highly Popular"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Determine the shape of our training and testing sets.
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Splitting into Train and Test sets into an 80/20 split.
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, random_state=78, train_size=0.80)
```

```python
# Determine the shape of our training and testing sets.
print(X_train2.shape)
print(X_test2.shape)
print(y_train2.shape)
print(y_test2.shape)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train2)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train2)
X_test_scaled = X_scaler.transform(X_test2)
```

```python
# Creating the decision tree classifier instance.
model = tree.DecisionTreeClassifier()
# Fitting the model.
model = model.fit(X_train_scaled, y_train2)
```

```python
# Making predictions using the testing data.
predictions = model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test2, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test2, predictions)
```

```python
# Displaying results
print("Confusion Matrix - Decision Tree Classifier")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test2, predictions))
```
